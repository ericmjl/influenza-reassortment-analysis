# Background

Turns out there were some dropped nodes, so their host group wasn't coded up properly. I have found out which of those they are, and asked Nichola for help coding up them.

In this notebook, I will recombine them together into a single CSV file that can be used.

```python
import pandas as pd
import networkx as nx
import numpy as np
```

```python
G = nx.read_gpickle('20141103 All IRD Final Graph.pkl')
```

```python
# Get a set of 'country:host species' mappings:

country_host = set()
for n, d in G.nodes(data=True):
    country = d['country']
    host = d['host_species']
    country_host.add((country, host))
    
country_host_new = pd.DataFrame(list(country_host))
country_host_new.columns = ['Country', 'Species']
country_host_new
```

```python
country_host_old = pd.read_csv('Country-Species_to_be_Coded_by_Nichola.csv', index_col=0)
country_host_old
```

```python
country_host_comb = pd.merge(country_host_new, country_host_old, on=['Country', 'Species'], how='outer')
country_host_comb[country_host_comb['Habitat/setting'].isnull()].to_csv('Country-Species_to_be_coded_replenished.csv')
```

```python
country_host_comb = pd.read_csv('Country-Species_to_be_coded_replenished_NH.csv', index_col=0).sort()
country_host_comb
```

```python
master = pd.concat([country_host_comb, country_host_old]).reset_index(drop=True)
master.to_csv('Country-Species_coded.csv')
master
```
