# Introduction

To investigate this question, I will plot a scatterplot of time delta vs. PWI for every edge.

```python
import networkx as nx
import matplotlib.pyplot as plt
import time

from datetime import datetime

%matplotlib inline
```

```python
G = nx.read_gpickle('20141103 All IRD Final Graph.pkl')
```

```python
tds_all = []
pwis_all = []

tds_reassortant = []
pwis_reassortant = []

for sc, sk, d in G.edges(data=True):
    sc_time = G.node[sc]['collection_date']
    sk_time = G.node[sk]['collection_date']
    
    td = sk_time - sc_time
    if d['edge_type'] == 'full_complement':
        tds_all.append(td.days)
        pwis_all.append(d['pwi'])
        
    if d['edge_type'] == 'reassortant':        
        tds_reassortant.append(td.days)
        pwis_reassortant.append(d['pwi'])
```

```python
plt.scatter(tds_all, pwis_all, alpha=1, color='blue', label='Full Complement')
plt.ylabel('PWI')
plt.xlabel('Time Delta (Days)')
plt.title('PWIs vs. Time Deltas (Full Complement)')
plt.ylim(7.1, 8.1)
plt.xlim(-2000, 12000)
# plt.legend()
plt.savefig('PWIs vs. Time Deltas (Full Complement).pdf', bbox_inches='tight')
```

```python
plt.hist2d(tds_all, pwis_all, bins=20, cmap='Blues')
plt.ylim(7.1, 8.1)
plt.xlim(-2000, 12000)
plt.show()
```

```python
plt.scatter(tds_reassortant, pwis_reassortant, alpha=1, color='green', label='Reassortant')
plt.ylabel('PWI')
plt.xlabel('Time Delta (Days)')
plt.title('PWIs vs. Time Deltas (Reassortant)')
plt.savefig('PWIs vs. Time Deltas (Reassortant).pdf', bbox_inches='tight')
```

```python
plt.hist2d(tds_reassortant, pwis_reassortant, bins=20, cmap='Greens')
plt.ylim(7.1, 8.1)
plt.xlim(-2000, 12000)
plt.show()
```

Most of the edges found are short time delta (within a few months) and of high PWI (less than 0.05 average genetic diversity persegment).

```python
# Just to check: Is time delta in days related to the collection date for the source isolate?

sk_dates = []
sc_dates = []
time_deltas = []

for sc, sk, d in G.edges(data=True):
    if d['edge_type'] == 'reassortant':
        sc_date = G.node[sc]['collection_date']
        sk_date = G.node[sk]['collection_date']
        
        td = sk_date - sc_date
        
        sk_dates.append(sk_date.toordinal())
        sc_dates.append(sc_date.toordinal())
        time_deltas.append(td.days)
```

```python
plt.scatter(sk_dates, time_deltas, color='green', alpha=0.05)
plt.xlabel('Sink Isolate Date')
plt.ylabel('Edge Time Delta')
plt.title("Reassortant Edges' Time Delta vs. \n Sink Collection Date")
plt.show()
```

```python
plt.scatter(sc_dates, time_deltas, color='green', alpha=0.05)
plt.xlabel('Source Isolate Date')
plt.ylabel('Edge Time Delta')
plt.title("Reassortant Edges' Time Delta vs. \n Source Collection Date")
plt.show()
```
